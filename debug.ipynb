```python
import pandas as pd
```

```python
equity_df = pd.read_csv('utils\\ibov_equity.csv', sep=';')
```

```python
index_df = pd.read_csv('utils\\ibov_equity_index.csv', sep=';')
```

```python
index_df = index_df.drop(columns=['override', 'override_period'])
index_df['date'] = pd.to_datetime(index_df['date'])
index_df['value'] = index_df['value'].str.replace(',', '').astype(float)
index_df['extraction_date'] = pd.to_datetime(index_df['extraction_date'])
index_df
```

```python
PX_Last_index_df = index_df[index_df['field'] == 'PX_Last'].sort_values('extraction_date').drop_duplicates(subset=['date', 'ticker', 'field'], keep='last').sort_values(['ticker', 'date']).set_index('date', drop=True).copy()
CUR_MKT_CAP_index_df = index_df[index_df['field'] == 'CUR_MKT_CAP'].sort_values('extraction_date').drop_duplicates(subset=['date', 'ticker', 'field'], keep='last').sort_values(['ticker', 'date']).set_index('date', drop=True).copy()
```

```python
equity_df['date'] = pd.to_datetime(equity_df['date'])
equity_df['value'] = equity_df['value'].str.replace(',', '').astype(float)
equity_df['extraction_date'] = pd.to_datetime(equity_df['extraction_date'])
```

```python
PX_Last_df = equity_df[equity_df['field'] == 'PX_Last'].copy()
PX_Last_df = PX_Last_df.sort_values('extraction_date').drop_duplicates(subset=['date', 'ticker', 'field'], keep='last').sort_values(['ticker', 'date']).set_index('date', drop=True).copy()
PX_Last_df.tail()
```

```python
CUR_MKT_CAP_df = equity_df[equity_df['field'] == 'CUR_MKT_CAP'].copy()
CUR_MKT_CAP_df = CUR_MKT_CAP_df.sort_values('extraction_date').drop_duplicates(subset=['date', 'ticker', 'field'], keep='last').sort_values(['ticker', 'date']).set_index('date', drop=True).copy()
CUR_MKT_CAP_df.tail()
```

```python
PX_Last_df = pd.concat([PX_Last_df, PX_Last_index_df], axis=0)
CUR_MKT_CAP_df = pd.concat([CUR_MKT_CAP_df, CUR_MKT_CAP_index_df], axis=0)
```

```python
PX_Last_pivot_df = PX_Last_df.pivot(columns='ticker', values='value')
PX_Last_pivot_df.tail()
```

```python
CUR_MKT_CAP_pivot_df = CUR_MKT_CAP_df.pivot(columns='ticker', values='value')
CUR_MKT_CAP_pivot_df.tail()
```

```python
# PX_Last_df: PX_Last in long format.
# CUR_MKT_CAP_df: CUR_MKT_CAP in long format.
# PX_Last_pivot_df: PX_Last in wide format.
# CUR_MKT_CAP_pivot_df: CUR_MKT_CAP in wide format.
```
